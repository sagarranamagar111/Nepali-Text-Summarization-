```python
# Install required libraries
!pip install datasets --q
import pandas as pd
from datasets import load_dataset
from sklearn.model_selection import train_test_split
```

```python
# Load the XLSUM dataset for Nepali language
xl_sum = load_dataset("GEM/xlsum", "nepali")

# Convert each split of the XLSUM dataset to a DataFrame
xl_sum_train_df = pd.DataFrame(xl_sum["train"])
xl_sum_validation_df = pd.DataFrame(xl_sum["validation"])
xl_sum_test_df = pd.DataFrame(xl_sum["test"])

# Add an 'id' column to each split of the XLSUM dataset
xl_sum_train_df['id'] = xl_sum_train_df.index.astype(str)
xl_sum_validation_df['id'] = xl_sum_validation_df.index.astype(str)
xl_sum_test_df['id'] = xl_sum_test_df.index.astype(str)

# Save the XLSUM dataset to a CSV file
xl_sum_train_df.to_csv("/content/drive/My Drive/FinalFinal/xlsum_train.csv", index=False)
xl_sum_validation_df.to_csv("/content/drive/My Drive/FinalFinal/xlsum_validation.csv", index=False)
xl_sum_test_df.to_csv("/content/drive/My Drive/FinalFinal/xlsum_test.csv", index=False)

# Combine the splits into one DataFrame for XLSUM
xl_sum_df = pd.concat([xl_sum_train_df, xl_sum_validation_df, xl_sum_test_df])
```

```python
# Load the Nepali News dataset
news_dataset = load_dataset("wisewizer/nepali-news")

# Convert the entire Nepali News dataset to a DataFrame by concatenating train, validation, and test splits
news_df = pd.concat([pd.DataFrame(news_dataset[split]) for split in news_dataset])

# Remove all columns except 'id', 'text', and 'summary'
news_df = news_df[['id', 'text', 'summary']]

# Randomly delete 15% of the data from the Nepali News dataset
news_df = news_df.sample(frac=0.85, random_state=42)
```

```python
# Split the Nepali News dataset into train, test, and validation sets
train_df, temp_df = train_test_split(news_df, test_size=0.3, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# Save the split datasets to CSV files in Google Drive
train_df.to_csv('/content/drive/My Drive/FinalFinal/news_train.csv', index=False)
val_df.to_csv('/content/drive/My Drive/FinalFinal/news_val.csv', index=False)
test_df.to_csv('/content/drive/My Drive/FinalFinal/news_test.csv', index=False)
```

```python
# Combine the XLSUM and Nepali News datasets with only 'id', 'text', and 'summary' columns
final_dataset = pd.concat([
    xl_sum_df[['id', 'text', 'target']].rename(columns={'target': 'summary'}),
    news_df[['id', 'text', 'summary']]
])

# Save the final dataset to a CSV file
final_dataset.to_csv('final_dataset.csv', index=False)
final_dataset
```
